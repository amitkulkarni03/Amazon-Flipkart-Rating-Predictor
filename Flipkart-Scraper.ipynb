## Import Required Packages

```python
import pandas as pd
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

## Create list of products

```python
products=['laptops', 'Phones', 'Headphones', 'smart watches', 'Cameras', 'Printers', 'monitors', 'Home theater', 'router']
```

```python
products
```

## get URLs

```python
s = Service(r'Chromedriver.exe')
driver = webdriver.Chrome(service=s)
driver.get('https://www.flipkart.com/')
driver.find_element(By.XPATH,'//div[@class="_2QfC02"]/button[contains(text(), "✕")]').click()

flip_urls = []
for product in products:
    # fetch search box and insert our list
    search_box = driver.find_element(By.XPATH,'//input[@class="_3704LK"]')
    search_box.clear()
    search_box.send_keys(product)
    # click on search button
    driver.find_element(By.XPATH,'//button[@class="L0Z3Pu"]').click()
    time.sleep(2)

    urls = []
    url_tags = []
    j = 0
    while j < 11:
        try: 
            url_tags = driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
        except: 
            url_tags = driver.find_elements(By.XPATH,'//a[@class="_2rpwqI"]')
        for tag in url_tags:
            urls.append(tag.get_attribute('href'))
        try:
            driver.find_elements(By.XPATH,'//a[@class="_1LKTO3"]')[-1].click()
            time.sleep(2)
        except:
            break
        j += 1
    
    flip_urls.extend(urls)

```

## get stars and reviews for URLs

```python
flip_stars = []
flip_reviews = []
s = Service(r'Chromedriver.exe')
driver = webdriver.Chrome(service=s)

for url in flip_urls:
    driver.get(url)
    stars = []
    reviews = []
        
    try:
        # expand the 'all reviews' section
        driver.find_element(By.XPATH,'//div[@class="_3UAT2v _16PBlm"]').click()
        time.sleep(2)
        
        # scrape star ratings
        try:
            star_tags = driver.find_elements(By.XPATH,'//div[contains(@class, "_3LWZlK")]')
            for i in range(1, len(star_tags)):
                stars.append(star_tags[i].text)
        except Exception as e:
            print(f"Error in scraping star ratings for {url}: {e}")
        
        # scrape reviews
        try:
            review_tags = driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]/div/div')
            for tag in review_tags:
                reviews.append(tag.text)
        except Exception as e:
            print(f"Error in scraping reviews for {url}: {e}")
        
        flip_stars.extend(stars)
        flip_reviews.extend(reviews)
    except Exception as e:
        print(f"Error in scraping data for {url}: {e}")
        
driver.quit()

print(flip_stars)
print(flip_reviews)
```
