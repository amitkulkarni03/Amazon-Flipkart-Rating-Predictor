## Import Required Packages

```python
import pandas as pd
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

## Create list of products

```python
products=['laptops', 'Phones', 'Headphones', 'smart watches', 'Cameras', 'Printers', 'monitors', 'Home theater', 'router']
```

```python
products
```

## get URLs

```python
s=Service(r'Chromedriver.exe')
driver=webdriver.Chrome(service=s)
driver.get('https://www.amazon.in/')
amazon_urls=[]
for product in products :
    driver.implicitly_wait(50)
    #fetch search box and insert our list
    driver.find_element(By.XPATH,'//input[@id="twotabsearchtextbox"]').send_keys(product)
    #click on search button
    driver.find_element(By.XPATH,'//input[@id="nav-search-submit-button"]').click()
    j=0
    urls=[]
    url_tags=[]
    while j<11:
        url_tags=driver.find_elements(By.XPATH,'//a[@class="a-link-normal a-text-normal"]')
        for tag in url_tags:
            urls.append(tag.get_attribute('href'))
        driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]').click()
        j+=1
    amazon_urls.append(urls)
    driver.find_element(By.XPATH,'//input[@id="twotabsearchtextbox"]').clear()
    
```

```python
urls=[]
for i in range(0,9):
    for url in amazon_urls[i]:
        urls.append(url)
```

```python
len(urls)
```

## get stars and reviews for URLs

```python
star=[]
reviews=[]
for url in urls:
    driver.get(url)
    driver.implicitly_wait(10)
    try:
        star_tags=driver.find_elements(By.XPATH,'//div[@id="cm-cr-dp-review-list"]/div/div/div/div[2]/a/i/span')
        for tag in star_tags:
            star.append(tag.get_attribute('textContent'))
    except:
            star.append('-')
    try:
        review_tags=driver.find_elements(By.XPATH,'//div[@id="cm-cr-dp-review-list"]/div/div/div/div[4]')
        for tag in review_tags:
            reviews.append(tag.get_attribute('textContent'))
    except:
            reviews.append('-')
```

## Create DataFrame and CSV file

```python
df_amazonReviews=pd.DataFrame({})
```

```python
df_amazonReviews['star']=star
df_amazonReviews['reviews']=reviews
```
